```python
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn import preprocessing, model_selection
from sklearn.cluster import MeanShift
style.use('ggplot')
```

```python
df = pd.read_excel("titanic.xls")
original_df = pd.DataFrame.copy(df) 
#we can't just assign a dataframe to another dataframe, then changes in df will also take place in original_df
df.head()
```

```python
df.drop(columns =["body","name"],inplace =True)
```

```python
df.apply(pd.to_numeric, errors="ignore")
df.fillna(0,inplace =True)
df.head()
```

```python
print(df.columns)
```

```python
def handle_non_numeric_data(df):
    for column in df.columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]
        if df[column].dtype != np.int64 or df[column].dtype != np.float64:
            column_contents = df[column].tolist()
            unique_elements = set(column_contents)
            x=0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x+=1
            df[column] = list(map(convert_to_int, df[column]))
    return df
```

```python
df = handle_non_numeric_data(df)
df.head()
```

```python
X = np.array(df.drop(columns="survived"))
X = preprocessing.scale(X)
y = np.array(df["survived"])

clf = MeanShift()
clf.fit(X)
```
